```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = pd.read_csv('tip.csv')
titanic = pd.read_csv('Titanic-Dataset.csv')
flights = pd.read_csv('flights.csv')
iris = pd.read_csv('iris.csv')
```

```python
# numerical numerical
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', size='size', data=tips)
```

```python
# Numerical Categorical
sns.barplot(x='Pclass', y='Fare', hue='Sex', data=titanic)
```

```python
# Box plot categorical numerical
sns.boxplot(x='Sex', y='Age',hue='Survived', data=titanic)
```

```python
# heatmap categorical categorical
df = pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(df)
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```
